## Data cleaning

```python
# Read CSV file

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\User\\Desktop\\python\\github\\kdrama.csv' )
pd.set_option('display.max_columns', None)
```

```python
# Let's see how the records look like. 

df.head()
```

```python
# Let's get some info of the data frame.

df.info()
```

Convert columns `Original Network`, `Aired On`, `Genre`, `Tags`, `Cast` and `Director`, `Screenwriter`, `Production companies` (the last three columns have NaN) from object type to list.

```python
cols = ['Original Network', 'Aired On', 'Genre', 'Tags', 'Cast', 'Director', 'Screenwriter', 'Production companies']
df[cols] = df[cols].fillna('').apply(lambda col: col.str.split(',').apply(lambda x: [i.strip() for i in x]))

df[cols].head()
```

Convert `Aired Date` column into two columns `Start Date` and `Finish Date`. If all the episodes have been released the same day, then `Start Date` and `Finish Date` are the same.

```python
# Split 'Aired Date' into two columns 'Start Date' and 'Finish Date'
df[['Start Date', 'Finish Date']] = df['Aired Date'].str.split(' - ', expand=True)

# Convert both columns to datetime
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce', format='%b %d, %Y')
df['Finish Date'] = pd.to_datetime(df['Finish Date'], errors='coerce', format='%b %d, %Y')

# If 'Finish Date' is NaN (only one date was provided), set it equal to 'Start Date'
df['Finish Date'] = df['Finish Date'].fillna(df['Start Date'])

print(df[['Start Date', 'Finish Date']])
```

Convert column `Duration` from object type to timedelta.

```python
# Extract hours and minutes
df[['Hours', 'Minutes']] = df['Duration'].str.extract(r'(?:(\d+) hr\.)?\s*(\d+) min\.')

# Fill NaN hours with 0 and convert to integers
df['Hours'] = df['Hours'].fillna(0).astype(int)
df['Minutes'] = df['Minutes'].astype(int)

# Convert to timedelta
df['Duration'] = pd.to_timedelta(df['Hours'], unit='h') + pd.to_timedelta(df['Minutes'], unit='m')

# Delete unnecessary columns
df.drop(['Aired Date','Hours', 'Minutes', 'Rank'], axis=1, inplace=True)

print(df.head())
```

Fill in missing `Content Ratings` where the info can be found on the Internet.

```python
# Check what are the values in column Content Rating.

print(df['Content Rating'].unique())

# Filter the rows where 'Content Rating' column is NaN.

filtered_df = df[df['Content Rating'].isna()]
filtered_df
```

```python
# Change the values in 'Content Rating' column with info found on the Internet.

df.loc[140, 'Content Rating'] = '13+ - Teens 13 or older'
df.loc[171, 'Content Rating'] = '15+ - Teens 15 or older'
```

## Data analysis

Add a column calculating total length of kdrama.

```python
df['Total Duration'] = df['Number of Episodes'] * df['Duration']
```

Add a column describing how many days was kdrama aired.

```python
df['Air Days'] = (df['Finish Date'] - df['Start Date']).dt.days
```

```python
df.head()
```

```python
# Summary statistics for numerical columns

df.describe()
```

1. What is the the distribution of kdramas by year?

```python
kdramas_per_year = df['Year of release'].value_counts().sort_index()

kdramas_per_year.plot(kind='bar', color='lightcoral')
plt.title('Kdramas per year')
plt.xlabel(None)
plt.show()
```

2. Which years have the most high-rated kdramas released?

```python
df['Year of release'].value_counts().head()
```

3. What is the average rating of kdramas per year?

```python
df.groupby('Year of release')['Rating'].mean().sort_values(ascending=False)
```

4. Which dramas have the highest ratings?

```python
df[['Name', 'Rating']].sort_values(by='Rating', ascending=False).head(10)
```

5. What is the best kdrama in every `Genre`?

```python
df_exploded = df.explode('Genre')

first_kdramas_per_genre = df_exploded.groupby('Genre', as_index=False).first()
first_kdramas_per_genre = first_kdramas_per_genre[['Genre', 'Name', 'Year of release', 'Number of Episodes', 'Total Duration']]
first_kdramas_per_genre
```

6. Who are the most frequently casted actors?

```python
df.explode('Cast')['Cast'].value_counts().head(10)
```

7. On which days are kdramas most frequently aired?

```python
day_counts = df.explode('Aired On')['Aired On'].value_counts()
print(day_counts)

# Sort by weekday order
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
day_counts = day_counts.reindex(day_order)

plt.figure(figsize=(8, 5))
day_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel(None)
plt.title("Number of kdramas aired on each day of the week")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Grouped by each set of days
df['Airing Schedule'] = df['Aired On'].apply(tuple) # list is an unhashable type, so need to change to tuple
df.groupby('Airing Schedule')['Name'].count().sort_values(ascending=False).head(5)
```

8. Which days of the week are dramas released?

```python
df['Release Day'] = df['Start Date'].dt.day_name()
release_day = df['Release Day'].value_counts()

release_day.plot(kind='barh', cmap='viridis')
plt.title('Number of Kdramas by Release Day')
plt.ylabel(None)
plt.show()
```
